```python
import matplotlib.pyplot as plt
import pandas as pd
import pickle
```

```python
# read csv file
train_df = pd.read_csv('./info/train_patients_info.csv')
```

```python
# Count the occurrences of each group for each BMI category
group_counts = train_df.groupby(['BMI_Category', 'Group']).size().unstack(fill_value=0)

new_index = ['Underweight', 'Healthy', 'Overweight', 'Obesity']
group_counts = group_counts.reindex(new_index)

# Create a bar plot to visualize the counts
ax = group_counts.plot(kind='bar', stacked=False, figsize=(10, 6))

# Add labels and a title
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.title('Disease Distribution for each BMI Category - Train')

# Show the plot
plt.legend(title='Group')
plt.show()
```
